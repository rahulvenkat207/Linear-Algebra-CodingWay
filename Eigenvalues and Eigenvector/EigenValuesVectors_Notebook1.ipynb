```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Define the matrix A
A = np.array([[1, 6], [5, 2]])

# Define the vectors u and v
u = np.array([6, -5])
v = np.array([3, -2])
```

```python
# Calculate A * u and A * v
Au = np.dot(A, u)
Av = np.dot(A, v)

# Check if Au and Av are scalar multiples of u and v
lambda_u = Au[0] / u[0]  # Eigenvalue associated with u
lambda_v = Av[0] / v[0]  # Eigenvalue associated with v
print(lambda_u)
print(lambda_v)
```

```python
# Check if the eigenvalue associated with u is the same for both vectors
if np.allclose(lambda_u, lambda_v):
    print("Both vectors u and v are eigenvectors of A with eigenvalue:", lambda_u)
else:
    print("Vectors u and v are not eigenvectors of A.")
```

```python
# Print the results
print("A * u =", Au)
print("A * v =", Av)
```

```python
def plot_eigenvectors(u, v, Au, Av):
    plt.figure(figsize=(8, 6))

    # Plot the vectors u and v
    plt.plot([0, u[0]], [0, u[1]], 'r-', label='u')
    plt.plot([0, v[0]], [0, v[1]], 'b-', label='v')

    # Plot the transformed vectors A * u and A * v
    plt.plot([0, Au[0]], [0, Au[1]], 'r--', label='A * u')
    plt.plot([0, Av[0]], [0, Av[1]], 'b--', label='A * v')

    # Add names to the vectors in legend
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Visualization of Eigenvectors')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend(loc='upper right')
    plt.axis('equal')
    plt.show()

# Example usage:
plot_eigenvectors(u, v, Au, Av)
```
