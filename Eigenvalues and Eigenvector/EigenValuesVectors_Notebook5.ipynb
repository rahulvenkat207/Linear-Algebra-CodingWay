**Transition Matrix Definition**: The program starts by defining a transition matrix. This matrix represents the probabilities of transitioning from one page to other page.

Initialization of Initial State **bold text**: The initial state vector x is defined. This vector represents the probabilities of starting from each page(intially).

**Matrix-Vector Multiplication Function**: The program defines a function called matrix_vector_multiply to perform matrix-vector multiplication. This function takes a matrix and a vector as input and returns the result of the multiplication.

Result Output **bold text**: After 100 iterations, the composition of the resulting vector x is printed.


# Define the transition matrix

```python
transition_matrix = [
    [0, 0, 1, 0.5],
    [1/3, 0, 0, 0],
    [1/3, 0.5, 0, 0.5],
    [1/3, 0.5, 0, 0]
]
```

```python
def matrix_vector_multiply(matrix, vector):
    """
    Perform matrix-vector multiplication.

    Args:
    - matrix (list of lists): The matrix to be multiplied.
    - vector (list): The vector to be multiplied.

    Returns:
    - result (list): The result of the matrix-vector multiplication.
    """
    result = [0] * len(matrix)
    for i in range(len(matrix)):
        for j in range(len(vector)):
            result[i] += matrix[i][j] * vector[j]
    return result
```

Lets say you start on Page 1, and then depending on this matrix, you move to the next page. Repeat this process, say
a hundred times. How would the composition of vector x look ?

```python
# Define the initial vector x (start on Page 3)
x = [1, 0, 0, 0]

# Perform matrix-vector multiplication 100 times
for _ in range(100):
    # Perform matrix-vector multiplication
    x = matrix_vector_multiply(transition_matrix, x)

# Print the composition of vector x after 100 iterations
print("Composition of vector x after 100 iterations starting from Page 1:")
for i, prob in enumerate(x):
    print(f"Page {i+1}: {prob}")
```

Lets say you start on Page 3, and then depending on this matrix, you move to the next page. Repeat this process, say
a hundred times. How would the composition of vector x look ?

```python
# Define the initial vector x (start on Page 3)
x = [0, 0, 1, 0]

# Perform matrix-vector multiplication 100 times
for _ in range(100):
    # Perform matrix-vector multiplication
    x = matrix_vector_multiply(transition_matrix, x)

# Print the composition of vector x after 100 iterations
print("Composition of vector x after 100 iterations starting from Page 3:")
for i, prob in enumerate(x):
    print(f"Page {i+1}: {prob}")
```


x after 100 iterations starting from both Page 1 and Page 3 is the same, it means that regardless of the starting page, the system reaches the same steady-state distribution.
