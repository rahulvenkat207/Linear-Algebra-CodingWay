T:R2 -> R2represents a linear transformation from a two-dimensional vector space (R2 ) to another two-dimensional vector space (R2)

```python
import matplotlib.pyplot as plt
import numpy as np
```

A basis vector is a vector that spans a space, meaning any other vector in that space can be represented as a linear combination of the basis vectors.

A basis should be

i)squared matrix
ii)Linearly independent


# Given Basis vectors

```python
e1 = np.array([[1], [0]])
e2 = np.array([[0], [1]])

print("Basis vector e1:\n ",e1)
print("Basis vector e2:\n",e2)
```

# plot these basis vectors

```python
# Plot basis vectors
plt.quiver(0, 0, e1[0,0], e1[1,0], angles='xy', scale_units='xy', scale=1, color='r', label='e1')
plt.quiver(0, 0, e2[0,0], e2[1,0], angles='xy', scale_units='xy', scale=1, color='b', label='e2')

plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Basis Vectors e1 and e2')
plt.grid()
plt.legend()
plt.show()
```

# Given Matrix

```python
# Define diagonal matrix D
D = np.array([[3, 0], [0, 6]])
print(D)
```

# transform these basis vectors using the matrix D

```python
# Transform basis vectors using D
e1_transformed = np.dot(D, e1)
e2_transformed = np.dot(D, e2)
print("Transformed e1:\n",e1_transformed)
print("Transformed e2\n",e2_transformed)
```

```python
# Plot transformed basis vectors
plt.quiver(0, 0, e1_transformed[0], e1_transformed[1], angles='xy', scale_units='xy', scale=1, color='r', label='e1 transformed')
plt.quiver(0, 0, e2_transformed[0], e2_transformed[1], angles='xy', scale_units='xy', scale=1, color='b', label='e2 transformed')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Transformed Basis Vectors')
plt.grid()
plt.legend()
plt.show()
```

This means

e1 gets stretched by a factor of 3 along the x-axis and

e2 gets stretched by a factor of 6 along the y-axis.

# transform the vector x = [1, 1] using the diagonal matrix D:

```python
x = [[1], [1]]
x_transformed = np.dot(D, x)
print(x_transformed)
```

```python
# Plot original and transformed vectors
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='g', label='x')
plt.quiver(0, 0, x_transformed[0], x_transformed[1], angles='xy', scale_units='xy', scale=1, color='m', label='D * x')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Transformation of Vector x')
plt.grid()
plt.legend()
plt.show()
```

x gets stretched by a factor of 3 along the x-axis and by a factor of 6 along the y-axis

```python
D_squared = np.dot(D, D)
print("D squared:")
print(D_squared)
```

When we multiply a diagonal matrix by itself (or by another diagonal matrix), the result is still a diagonal matrix,meaning that performing operations such as squaring a diagonal matrix or raising it to any positive integer power results in another diagonal matrix.
